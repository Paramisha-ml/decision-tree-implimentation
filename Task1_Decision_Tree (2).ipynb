```python
import pandas as pd
import matplotlib as plt #for visulization
```

```python
path="/content/drive/MyDrive/dataset/1_JobPrediction.csv" #path of the exal dataset that i have prepared
df = pd.read_csv(path)
df.head(13) #this is the actual lenght of the complete table
```

```python
d = {'Bachelors': 0, 'Masters': 1, 'phD': 2} #dictonary for assigning numeric values
df['DEGREE'] = df['DEGREE'].map(d)  #changing the column value with numeric values
d = {'Yes': 1, 'No': 0}
df['HIRED'] = df['HIRED'].map(d)

print(df.head(13))

```

```python
features = ['AGE', 'EXPERIENCE', 'SCORE', 'DEGREE']
X = df[features]
Y = df['HIRED']
combine = pd.concat([X, Y],axis=1) #this is for combaing the input and output table together
print(combine)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

tree = DecisionTreeClassifier()
tree = tree.fit(X, Y)

plot_tree(tree, feature_names=features)

```
