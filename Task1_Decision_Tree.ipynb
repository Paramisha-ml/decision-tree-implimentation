```python
import pandas as pd
import matplotlib as plt # for creating tree visulization
```

```python
data = pd.read_csv('/content/drive/MyDrive/dataset /hiring_dataset_100_rows.csv')# loding the dataset
data.head()
```

```python
d = {'Bachelors': 0, 'Masters': 1, 'phD': 2} #so degree column is texual column so we are creating a dictonary for assigning numeric values
data['DEGREE'] = data['DEGREE'].map(d)  #changing the column value with numeric values
d = {'Yes': 1, 'No': 0}
data['HIRED'] = data['HIRED'].map(d)

print(data.head())
```

```python
features = ['AGE', 'EXPERIENCE', 'CGPA', 'DEGREE']
X = data[features] #input featurs
Y = data['HIRED'] # output column
combine = pd.concat([X, Y],axis=1) #this is for combaing the input and output table together
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree # using scikit learn to create and tran the model

tree = DecisionTreeClassifier()
tree = tree.fit(X, Y) # traning the model

plot_tree(tree, feature_names=features) # this is for creating the final tree
```
